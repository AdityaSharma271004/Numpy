Advance operation with Business examples

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 170000, 190000, 210000], #Paradise Biryani
    [2, 130000, 180000, 150000, 250000], #Baba Biryani
    [3, 120000, 150000, 120000, 230000], #Chandu Biryani
    [4, 110000, 160000, 150000, 220000], #KFC Biryani
    [5, 160000, 120000, 100000, 210000], #Rakesh Biryani

])

print("==== zepto analysis ====")

print("\n sales data shape", sales_data.shape)

print("\n sample data for first 3 restaurant", sales_data[:3])

print("\n sample data for all", sales_data[:,1 :])


```

```python
# total sales per year row column

print(np.sum(sales_data, axis = 0))

yearly_sales = np.sum(sales_data[: ,1 :],axis = 1)
print(yearly_sales)

#minimum sales per year

print(np.min(sales_data[:,1:], axis = 1))


# maximum sales per year

max_sales = np.max(sales_data[:,1:], axis = 0)
print(max_sales)


#Average sales

avg_sales = np.mean(sales_data[:,1:],axis = 1)
print(avg_sales)


```

```python
#cumulative sales
cum_sum = np.cumsum(sales_data[:,1:],axis = 1)
print(cum_sum)
```
